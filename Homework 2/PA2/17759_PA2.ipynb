### Reward Values Plot

```python
%cd /content/gdrive/My Drive/MLGT/PA2
%cd maddpg/experiments
```

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

## MADDPG_simple
with open('../../results/maddpg_simple/test_rewards.pkl', 'rb') as f:
  rewards = pickle.load(f)
```

```python
plt.plot(np.arange(0, len(rewards)*1000, 1000), rewards, label='rewards')
plt.legend()
plt.xlabel('Episodes')
plt.xticks(np.arange(0, len(rewards)*1000, 1000))
plt.ylabel('Reward')
plt.show()
```

```python
## MADDPG_UAV
with open('../../results/maddpg_uav/test_agrewards.pkl', 'rb') as f:
  agent_rewards = pickle.load(f)
with open('../../results/maddpg_uav/test_rewards.pkl', 'rb') as f:
  mean_rewards = pickle.load(f)
```

```python
n_agent = len(mean_rewards)
agent_rewards = np.array(agent_rewards).reshape(n_agent, -1).transpose()
n_saves = len(agent_rewards[0])
```

```python
agent_rewards
```

```python
n_agent
```

```python
for i in range(3):
  plt.plot(np.arange(0, n_saves*1000, 1000), agent_rewards[i], label='Reward of Agent {}'.format(i))
plt.legend()
plt.xlabel('Episodes')
plt.xticks(np.arange(0, len(rewards)*1000, 1000))
plt.ylabel('Reward')
plt.show()
```

```python
## MADDPG_UAV
with open('../../results/maddpg_uav/test_agrewards.pkl', 'rb') as f:
  agent_rewards = pickle.load(f)
with open('../../results/maddpg_uav/test_rewards.pkl', 'rb') as f:
  mean_rewards = pickle.load(f)
```

```python
n_agent = len(mean_rewards)
agent_rewards = np.array(agent_rewards).reshape(n_agent, -1).transpose()
n_saves = len(agent_rewards[0])
```

```python
for i in range(3):
  plt.plot(np.arange(0, n_saves*1000, 1000), agent_rewards[i], label='Reward of Agent {}'.format(i))
plt.legend()
plt.xlabel('Episodes')
plt.xticks(np.arange(0, len(rewards)*1000, 1000))
plt.ylabel('Reward')
plt.show()
```
